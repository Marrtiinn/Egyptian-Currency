```python
import cv2
from ultralytics import YOLO

# el model
model = YOLO("best.pt") 

# el mapping lel class IDs lel values el 7a2e2ya
currency_mapping = {
    4: 100,  # 100 EGP (Face Front) 
    11: 100, # 100 EGP (Face Back) 
    8: 50,   # 50 EGP (Face Front) 
    2: 50,   # 50 EGP (Face Back) 
    6: 200,  # 200 EGP (Face Front) 
    7: 200,  # 200 EGP (Face Back) 
    9: 20,   # 20 EGP (Face Front) 
    5: 20,   # 20 EGP (Face Back) 
    10: 5,   # 5 EGP (Face Front) 
    1: 5,    # 5 EGP (Face Back) 
    0: 10,   # 10 EGP (Face Front)
    3: 10,  # 10 EGP (Face Back)
}

# el camera
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

print("Q for exit.")

# tazbeet el mapping w el display
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # man3 el false positives el low confidence w el overlapping
    results = model(frame, conf=0.6, iou=0.45, verbose=False)
    
    total_amount = 0

    # read el boxes w el class IDs w el confidence w el mapping lel values w el display
    for box in results[0].boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        class_id = int(box.cls[0])
        confidence = float(box.conf[0])

        #el results el 7a2e2ya
        coin_value = currency_mapping.get(class_id, 0)
        total_amount += coin_value

        # el box w el display
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        
        # ktabt el value w el confidence
        display_text = f"{coin_value} EGP ({confidence:.2f})"
        cv2.putText(frame, display_text, (x1, y1 - 10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # total display 
    cv2.rectangle(frame, (10, 10), (380, 70), (0, 0, 0), -1)
    cv2.putText(frame, f"Total: {total_amount} EGP", (20, 50), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)

    # el window
    cv2.imshow("Egyptian Currency Counter - YOLO11", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
